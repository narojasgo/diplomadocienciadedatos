#### <span style="color:#F3AA60"> Descripción del problema </span>

En términos generales, la teoría lingüística se ha construido principalmente a partir de las lenguas orales. Solo hasta la segunda mitad del siglo XX las lenguas de señas empezaron a ganar lugar en esta disciplina con la propuesta de análisis estructuras de William Stokoe (Oviedo, 2001). A esto, se suma una serie de dinámicas sociales y creencias erróneas sobre la sordera y la comunicación en las comunidades sordas que han dificultado el entendimiento adecuado de estas lenguas, como la educación oralizadora o la concepción de su comunicación como mímica. 

Por fortuna, los estudios lingüísticos de las últimas décadas y las luchas de la comunidad sorda han permitido establecer que las lenguas de señas son lenguas naturales capaces de expresar ideas complejas, suplen las necesidades comunicativas de sus hablantes, son fundamentales para el desarrollo cognitivo de los niños sordos y representan un factor cultural de gran importancia para la identidad de su comunidad de usuarios (Oviedo, 2001). De este modo, se reconoce que en la actualidad hay cerca de 300 lenguas de señas según [Naciones Unidas](https://www.un.org/en/observances/sign-languages-day#:~:text=Collectively%2C%20they%20use%20more%20than,informally%20when%20travelling%20and%20socializing.).

En el contexto nacional, la lengua de señas colombiana fue reconocida oficialmente el 1996 como la lengua propia de la comunidad sorda del país ([Caro y Cuervo](https://lenguasdecolombia.caroycuervo.gov.co/contenido/Lenguas-de-senas-colombiana/introduccion)). En la actualidad, su descripción lingüística todavía es limitado. Por ejemplo, no se cuenta con una gramática de la lengua; el trabajo más cercano a esto ha sido el de Oviedo en el libro Apuntes para una gramática de la lengua de señas colombiana (2001). Por lo tanto, existe una necesidad latente por consolidar investigaciones más profundas y extensas sobre esta lengua.

#### <span style="color:#F3AA60"> Fuente de datos </span>

Con el fin de responder a dicha necesidad, se creó el Semillero de investigación de lingüística de la lengua de señas colombiana al interior del pregrado en lingüística de la Universidad Nacional de Colombia. En este semillero, se recolectó un corpus léxico de la LSC, en otras palabras, un conjunto de señas que constituye un vocabulario de la LSC. Para esto, se sistematizaron 4583 señas encontradas en materiales disponibles en línea tales como el Diccionario básico de la LSC, libros de instituciones reconocidas, videos de vocabulario en youtube, etc. Así, se creó una base de datos en la que se compilaron todas las señas encontradas, registrándolas según su traducción en español. Ya que un mismo concepto (p. ej. "Perro") puede aparecer en diferentes materiales, la base de datos incluye cada una de las veces que apareció determinado concepto. Además, gracias al trabajo del equipo del semillero, también se identificó si esos registros son diferentes según su fonología (p.ej. el número de manos utilizados, la forma de la mano, el contacto con el cuerpo, etc.), por lo que un mismo concepto puede presentar variación léxica, es decir, distintas formas de decir el mismo concepto.

#### <span style="color:#F3AA60"> Variables </span> 

##### Base de variación léxica
1. Ocurrencia: número de identificación asignado y significado de la seña según la denominación que tenga en el material de donde fue tomado (fuente). (Categórica)
2. Concepto: generalización del nombre de la seña. (Categórica) 
3. Nuevo concepto sugerido: denominación modificada debido a que se evidenció un cambio de significado. (Categórica)
4. Alternativa automática: Columna concepto + Columna opción léxica. (Categórica) 
5. Revistar: opción para los analistas si consideraban que la seña debía ser discutida en grupo. (Categórica; bool)
6. Opción léxica: número asignado a la seña según la cantidad de formas para realizar un mismo concepto, siempre y cuando dichas formas fuesen distintas en dos o más parámetros fonológicos. Por defecto, se denominó 1 a la seña del DBSL.  (Categórica???)
7. Opción fonológica: letra (a-j) asignada a la seña cuando el registro cambiaba solo en un parámetro fonológico con respecto a la alternativa 1. (Categórica)
  8. Parámetro fonológico que varía: marcado solo en aquellas señas con opción fonológica b-j. (Categórica)
9. Observaciones: notas de la persona que describió, principalmente sobre otras características de la seña que se pueden analizar en otra ocasión. 
10. ¿Para discutir?: 
11. Nueva alternativa: Concepto + Opción léxica + Opción fonológica.
12. Nombre-vocabulario: nombre del material fuente.
13. Acceso-material-general: hipervínculo a la fuente en general.
14. Acceso-material-detalle: subespecificación de la ruta de acceso al material (ej. número de página).
15. Acceso-material-detalle-2: subespecificación de la ruta de acceso al material (ej. minuto del video).
16. Observaciones sobre concepto usado: notas de quien hizo la descripción. 
17. Género: marca de género gramatical identificada en la seña. 
18. Plural: marca de número gramatical identificada en la seña. 
19. Observaciones limpieza: notas sobre desiciones de la descripción y otras. 
20. Responsables-Limpieza-Concepto-Nueva Alternativa: persona que describió.
